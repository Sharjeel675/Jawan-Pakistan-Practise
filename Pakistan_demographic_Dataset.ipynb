```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV into pandas
data = pd.read_csv('Pakistan DemoGraphic/Age_Group_Gender_Distribution.csv')
df = pd.DataFrame(data)

# Selecting Age Group and Gender Data
Age = df['Age Group'].head(12)
Gender_Male = df['Male'].head(12)  # Negating Male values for population pyramid effect
Gender_Female = df['Female'].head(12)

# Figure Size
fig = plt.figure(figsize=(10, 7))

# Horizontal Bar Plot
plt.barh(Age, Gender_Male, color='blue', label='Male')
plt.barh(Age, Gender_Female, color='pink', label='Female')

# Labels and Title
plt.xlabel('Population')
plt.ylabel('Age Group')
plt.title('Age Group Gender Distribution')
plt.legend()

# Show Plot
plt.show()
```
