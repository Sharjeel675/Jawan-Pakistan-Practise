```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('house_sales.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
new_df = df.drop('price',axis = 'columns')
new_df
```

```python
model = linear_model.LinearRegression()
model.fit(new_df,df.price)
```

```python
model.predict([[3300]])
```

# Canada Per Capita Income Prediction

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.rename(columns={'per capita income (US$)': 'Capita_income'}, inplace=True)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Capita_income')
plt.scatter(df.year,df.Capita_income,color = 'red',marker = '+')
plt.show()
```

```python
new_df1 = df.drop('Capita_income',axis = 'columns')
new_df1
```

```python
model = linear_model.LinearRegression()
model.fit(new_df1,df.Capita_income)
```

```python
model.predict([[2022]])
```
