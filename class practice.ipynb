```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('factbook.db')
conn
```

```python
q = "select * from sqlite_master"
pd.read_sql_query(q, conn)
```

```python
q = "select * from facts"
pd.read_sql_query(q, conn)
```

```python
q = "select * from cities"
pd.read_sql_query(q, conn)
```

```python
q = '''
select f.name Country_Name , c.name City_Name,c.population City_Population from facts f
inner join cities c
ON
f.id = c.facts_id
where c.capital == 1 OR c.population > 10000000
 '''
     
pd.read_sql_query(q, conn)
```

```python
q = '''
select f.name Country_Name,c.name City_Name from facts f
inner join cities c
ON
f.id = c.facts_id
where f.name IS 'United Arab Emirates'
 '''
     
pd.read_sql_query(q, conn)
```

```python
q = '''
select c.population Capital_Population from facts f
inner join cities c
ON
f.id = c.facts_id
where c.capital == 0 AND c.population > 10000000 
order by Capital_Population desc
limit 3

 '''
     
pd.read_sql_query(q, conn)
```
