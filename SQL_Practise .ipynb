```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
sql_query('select * from sqlite_master')
```

```python
q = '''
select * from salaries
'''
df = sql_query(q)
df
```

# Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000

```python
q = '''
Select 
jobTitle,
OvertimePay
from Salaries
where 
OvertimePay > 50000

'''
sql_query(q)
```
