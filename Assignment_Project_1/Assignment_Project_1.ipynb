```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ABC.csv',encoding='latin1')
df
```

```python
df.head()
```

1. Identify the top 5 best-selling games globally.

```python
top_selling_game = df.nlargest(5,'Global')
print('Top 5 Best-Selling Game: ')
print(top_selling_game[['Game','Global']])
```

2. Compare regional sales (North America, Europe, Japan, and Rest of the World)
for each genre to understand market preferences.

```python
regional_sales = df.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
regional_sales['Total Sales'] = regional_sales.sum(axis=1)
print("Regional Sales by Genre:")
print(regional_sales)
```

3. Find the publisher with the highest total global sales.

```python
publisher_sales = df.groupby('Publisher')['Global'].sum()
top_publisher = publisher_sales.idxmax()
top_sales = publisher_sales.max()
print(f"The Top Publisher is {top_publisher}")
print(f"The Highest Sales is {top_sales}")
```

4. Analyze the year with the highest number of game releases.

```python
game_per_year = df.groupby('Year').size()
top_year = game_per_year.idxmax()
max_release = game_per_year.max()
print(f"The Heighest Number of Game release is: {top_year}")
print(f"Maximum per year game release is: {max_release}")
```

5. Determine the most popular genre globally based on total sales.

```python
genre_sale = df.groupby('Genre')['Global'].sum()
most_popular_genre = genre_sale.idxmax()
heighest_sale = genre_sale.max()
print(f"Most Popular Genre is: {most_popular_genre}")
print(f"Heighest Sales is: {heighest_sale}")
```

6. Identify the game with the highest sales in Japan compared to other regions.

```python
df['Other Region Sale'] = df['North America'] + df['Europe'] + df['Rest of World']
df['Japan Advantage'] = df['Japan'] - df['Other Region Sale']
top_game_japan = df.loc[df['Japan Advantage'].idxmax()]
print(f"Japan Sales: {top_game_japan['Japan']}")
print(f"Other Regions Sales: {top_game_japan['Other Region Sale']}")
print(f"Japan Advantage: {top_game_japan['Japan Advantage']:.2f}")
```

7. Examine if there is a correlation between sales in North America and Europe.

```python
correlation = df['North America'].corr(df['Europe'])
print(f"Correlation between North America and Europe sales: {correlation:.2f}")
```

8. Compare the sales growth trend for "Action" and "Shooter" genres over the
years.

```python
import matplotlib.pyplot as plt
filtered_data = df[df['Genre'].isin(['Action', 'Shooter'])]
sales_trend = filtered_data.groupby(['Year', 'Genre'])['Global'].sum().reset_index()
action_sales = sales_trend[sales_trend['Genre'] == 'Action']
shooter_sales = sales_trend[sales_trend['Genre'] == 'Shooter']
plt.figure(figsize=(8, 4))
plt.plot(action_sales['Year'], action_sales['Global'], label='Action', marker='o')
plt.plot(shooter_sales['Year'], shooter_sales['Global'], label='Shooter', marker='o')
plt.title('Sales Growth Trend for Action and Shooter Genres', fontsize=6)
plt.xlabel('Year', fontsize=6)
plt.ylabel('Total Global Sales (millions)', fontsize=6)
plt.legend()
plt.grid(True)
plt.show()
```
